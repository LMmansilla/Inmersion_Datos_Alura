```python
# Se visualizan los promedios incorporando la nueva columna que muestra el precio por metro cuadrado por barrio. 
inmuebles_ref3.groupby('Barrio').mean()
```

```python
# Calculo del promedio de precio por metro cuadrado por barrio, optimizada. 
# Paso 1: se realizan las sumas de los valores por barrio.

datos_barrio = inmuebles_ref3.groupby('Barrio').sum()
datos_barrio
```

```python
# Calculo del promedio de precio por metro cuadrado por barrio, optimizada. 
# Paso 2: se dividen las sumatorias de los valores por barrio.


datos_barrio['Valor_m2_Barrio'] = datos_barrio['Precio_Millon']/datos_barrio['Area']
datos_barrio
```

```python
# Calculo del promedio de precio por metro cuadrado por barrio, optimizada. 
# Paso 3: se genera un diccionario con los valores.

m2_barrio = dict(datos_barrio['Valor_m2_Barrio'])
m2_barrio
```

```python
inmuebles_ref3['Barrio']
```

```python
# Calculo del promedio de precio por metro cuadrado por barrio, optimizada. 
# Paso 5: se realiza un maps. En primera instancia se crea una columa con los datos de los barrios
# Luego a la misma se realiza un maps con el diccionario creado al DF que cuenta con todos los registros
# Con esto, todos los registros tendran una columna con el valor promedio del precio metro cuadrado de su barrio. 

inmuebles_ref3['Valor_m2_Barrio'] = inmuebles_ref3['Barrio']
inmuebles_ref3['Valor_m2_Barrio'] = inmuebles_ref3['Valor_m2_Barrio'].map(m2_barrio)
inmuebles_ref3.head(5)
     
```
