#### Top 10 del promedio de área de todos los inmuebles por barrio. 


```python
# Se realiza la limpieza de los outliers

# Crear una máscara booleana basada en la condición
condicion = inmuebles['Area'] < 10000

# Aplicar la máscara al DataFrame para filtrar las filas
inmuebles_ref = inmuebles[condicion]

inmuebles_ref.sort_values(by= 'Area', ascending=False).head(10)

```

```python
promedio_area_por_barrio = inmuebles_ref.groupby('Barrio')['Area'].mean().sort_values(ascending=False)
promedio_area_por_barrio.head(10).plot.bar()
```

```python
inmuebles_ref2 = inmuebles_ref
```

```python
inmuebles_ref2.sample(5)
```

```python
inmuebles_ref2.info()
```

```python
valor = inmuebles_ref2.Valor.str.split(expand=True)
valor
```

```python
serie_1 = valor[0]
serie_1

serie_2 = valor[1]
serie_2
```

```python
serie_1
```

```python
# Se agrega las 2 columnas al DF refinado

inmuebles_ref3 = inmuebles_ref2
inmuebles_ref3['Moneda'] = serie_1

inmuebles_ref3 = inmuebles_ref2
inmuebles_ref3['Precio'] = serie_2
```

```python
inmuebles_ref3.info()
```

```python
inmuebles_ref3['Precio'] = inmuebles_ref3['Precio'].str.replace('.','',regex=True)
```

```python
inmuebles_ref3
```

```python
inmuebles_ref3[['Precio','Barrio']]
```

```python
inmuebles_ref3['Precio_Millon'] = inmuebles_ref3.Precio.astype('float')/1000000
inmuebles_ref3.info()
```

```python
inmuebles_ref3
```

```python
pd.set_option('display.precision',2)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
inmuebles_ref3.describe()
```

```python
inmuebles_ref3.loc[inmuebles.Habitaciones == 110]
```

```python
inmuebles_ref3.loc[inmuebles.Area == 2]
```

```python

inmuebles_ref3['Precio_Millon'].plot.hist(bins=10)
```

```python
plt.hist(inmuebles_ref3['Precio_Millon'], bins=500, color='blue', alpha=0.7, edgecolor = "white")
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de Datos')
plt.xlim((50,1000))
plt.show()
```

```python

a = inmuebles_ref3.Precio_Millon < 300
b = inmuebles_ref3.Precio_Millon > 5

precios_meno = (a & b)
precios_meno
```

```python
precios_menor = inmuebles_ref3[precios_meno]
precios_menor
```

```python
precios_menor = precios_menor.reset_index()
```

```python
precios_menor.Precio_Millon
```

```python
import numpy as np
import seaborn as sns


sns.histplot(data=precios_menor, x='Precio_Millon', kde=True)
plt.xlabel('Precio (Millones)')
plt.ylabel('Frecuencia')
plt.title('Histograma de Precios')

# Muestra el gráfico
plt.show()
```
